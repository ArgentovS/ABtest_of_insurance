## 2.3. Формирование датасета застрахованных с действующими полюсами

```python
# Формируем запрос, объединяющий данные о полюсах и клиентах
# (убираем данные по заказам, в которых есть хотя бы одна не доставленная позиция)
query_0 = """
SELECT Value FROM currency_rate WHERE CharCode == 'USD'
"""
query_1 = """
            SELECT co.contract_id,
                   co.contract_num,
                   co.product_name,
                   cl.last_name, cl.first_name, cl.middle_name, cl.age, cl.sex, cl.client_id,
                   co.duration,
                   co.country,
                   co.price,
                   co.insurance_amount,
                   lo.loss_name, lo.loss_payout_amt,
                   co.currency_name
            FROM case_contracts co
            INNER JOIN case_clients cl
                    ON co.client_id = cl.client_id
            LEFT JOIN case_losses lo
                    ON co.client_id = lo.client_id
            WHERE co.contract_status == 'Действует'
        """
```

```python
# Заполняем датафрейм результатами запроса
connection = sqlite3.connect(f'{data_path}insurance')
print('База данных подключена')
# Загрузим датасет с заказами и временем доставки
CURRENCY_RATE = float(pd.read_sql_query(query_0, connection).Value[0].replace(',', '.'))
df = pd.read_sql_query(query_1, connection)
print('Датафрейм заполнен фичами и параметрами для расчёта целевой переменной')
connection.close()
print('Соединение с базой данных закрыто')
```

```python
# Проверим отсутствие пустот
df.info()
```

```python
# Проставим признак ошибки '-' в ячейки имени и отчества, которые не начинаются с буквы
# и объединим фамилию, имя и отчесвто в аббривеатуру
def chek_alph(x):
    if pd.isnull(x):
        return '-'
    elif x[0].isalpha():
        return x[0]
    else:
        return '-'
df['sum_name'] = ''
for column in ['last_name', 'first_name', 'middle_name']:
    df['sum_name'] = df['sum_name'] + df[column].apply(lambda x: chek_alph(x))
    df.drop(columns=column, inplace=True)
names = list(df['sum_name'])
df.insert(3, 'client_name', names)
df.drop(columns='sum_name', inplace=True)
```

## 2.4. Перерасчёт стоимости страховки в долларах

```python
# Функция перерасчёта
def dollar_bill(x, y):
    if x == 'Российский рубль':
        return int(y / CURRENCY_RATE)
    else:
        return int(y)
```

```python
# Пересчитаем стоимость в доллары
df['insurance_amount_$'] = df.apply(lambda x: dollar_bill(x.currency_name, x.insurance_amount), axis=1)
```

## 3.1. Заполнение оставшихся пропусков

```python
# Заполняем датафрейм результатами запроса
connection = sqlite3.connect(f'{data_path}insurance')
print('База данных подключена')
# Загрузим датасет с заказами и временем доставки
CURRENCY_RATE = float(pd.read_sql_query(query_0, connection).Value[0].replace(',', '.'))
df = pd.read_sql_query('SELECT * FROM result_table', connection)
print('Датафрейм заполнен фичами и параметрами для расчёта целевой переменной')
connection.close()
print('Соединение с базой данных закрыто')
```

**Справочно:** пропуски данных в фамилиях, именах и отчесвах были устранены в п.3.1 данного ноутбука при формировании единого признака "ФИО" из ячеек фималии, имени и отчества. Поскольку в страховых полюсах как в документе с финансовой ответсвенностью компании-страховщика ФИО заполняются по строгой форме - наличие таких пропусков может являться либо следствием техничесх сбоев при выгрузке данных из баз данных страхрвых компаний, либо указанием на аномальность записи, появившейся в базе данных страховых компаний как, например, тестовые записи при проверочных работах в информационной системе. В таких ячейках в случае пропусков данных или наличия не буквенных символов в замен признака проставлялся знак "-".

```python
# Проверим оставшиеся пропуски в датасете
df.isna().sum()
```

В датасете остались пропуски только в колонках `"наименование статьи убытка"` и `"сумма выплаты убытка"`. Данные статьи связаны друг с другом, а одинаковое количесвто пропуск в них подтверждает единую природу их возникновения. Учитывая это, с большой вероятностью, записи в данных двух признаках имеют пропуски в связи с тем что убытков по конкретным страховым полюсам не было. Таким образом данные признаки целесообразно заполнить следующими значениями:   
1. "-" в категрариальном признаке `loss_name`,
2. 0 в числовом признаке `loss_payout_amt`.

```python
# Заменим все путсые ячейки, относящие к выплатам по полюсу на значение 0,
# а в остальных ячейках на значение ' - '
df[['loss_name']] = df[['loss_name']].fillna(' - ')
df[['loss_payout_amt']] = df[['loss_payout_amt']].fillna(0)
```

## 3.2.2. Кодирование значений датасета для выявления аномалий

```python
# Проведём кодирование каждого признака 
df_fit_num = pd.DataFrame()
for column in df[check_columns].columns:
    le = LabelEncoder()
    df_fit_num[column] = le.fit_transform(df[column])
df_fit_num.head()
```

## 3.2.3. Определение выбросов взаимным дополнением результатов нескольких методов

```python
# Инициализируем словарь индексов записей-аномалей, в котором будем фиксировать
# метод выявления аномалий; признак с выявленной аномалией; список индексов записей-аномалий
indexs = {'Quantile': []}
```

```python
# Заполним словарь индексов записей-аномалий, индексами аномалий, выявленными квантилями более 3-сигма
for column in df_fit_num.columns:
    quantile_indexs = list(df_fit_num[df_fit_num[column] > df_fit_num[column].quantile(0.997)].index)
    if len(quantile_indexs) > 0:
        indexs['Quantile'].extend(quantile_indexs)
indexs['Quantile'] = list(set(indexs['Quantile']))
```

### 4.2.3. Кластаризация методом К-средних (с варацией количества оптимальных кластеров)

```python
kmeans = KMeans(n_clusters=10, n_init=10).fit(df_num_sc)
plt.scatter(df_num['X'], df_num['Y'], s=10, c=kmeans.labels_);
```

```python
# Определение оптимального количества кластеров
inertia = []
silhouette = []
N = 15
for k in range(2, N):
    kmeans = KMeans(n_clusters=k, random_state=1, n_init=10).fit(df_num_sc)
    preds = kmeans.predict(df_num_sc)
    inertia.append(np.sqrt(kmeans.inertia_))
    silhouette.append(silhouette_score(df_num_sc, preds))

# графическое отображение "колена", характеризующего оптимальное количесвто кластеров
plt.plot(range(2, N), silhouette, marker='s');
plt.xlabel('$k$')
plt.grid()
plt.ylabel('$J(C_k)$');
```

```python
# Первое ярковыраженное колено соответчствует 3-ём кластерам
kmeans = KMeans(n_clusters=3, n_init=10).fit(df_num_sc)
plt.scatter(df_num['X'], df_num['Y'], s=10, c=kmeans.labels_);
```

**Вывод 1:** Кластеризация методом K-средних графичесуки показывает наиболее чёткое выделение класеров чем Иерархическим методом.
